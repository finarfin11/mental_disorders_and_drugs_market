### Metadata

The Global organized crime index dataset has 193 observations of 33 features. It assesses levels of organized crime in 193 countries across the world and also their resilience to organized-criminal activity. Comprising two composite metrics, the Index ranks countries both according to their levels of criminality on a score from 1 to 10 and according to resilience to organized crime from 1 to 10 (lowest to highest levels). Full information and guiding questions can be viewed [here](../data/guiding_questions.pdf). The dataset provides information about levels of production, distribution, sale and consumption of heroin, cocaine, cannabis and synthetic drugs for 2021. Тhis data will be used for the purposes of the project.
 

The source of this dataset is https://ocindex.net/downloads.
